Вы работаете секретарем и вам постоянно приходят различные документы. Вы должны быть очень внимательны, чтобы не потерять ни один документ. Каталог документов хранится в следующем виде:

Перечень полок, на которых находятся документы хранится в следующем виде:

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

Необходимо реализовать пользовательские команды (внимание! это не название функций, которые должны быть выразительными, а команды, которые вводит пользователь, чтобы получить необходимый результат):
  
- p – people – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;  
- s – shelf – команда, которая спросит номер документа и выведет номер полки, на которой он находится  
*Правильно обработайте ситуации, когда пользователь будет вводить несуществующий документ*.
- l – list – команда, которая выведет список всех документов в формате   
**passport "2207 876234" "Василий Гупкин"**  
- as – add shelf – команда, которая спросит номер новой полки и добавит ее в перечень. *Предусмотрите случай, когда пользователь добавляет полку, которая уже существует*.
- d – delete – команда, которая спросит номер документа и удалит его из каталога и из перечня полок. *Предусмотрите сценарий, когда пользователь вводит несуществующий документ*;  
- m – move – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую. *Корректно обработайте кейсы, когда пользователь пытается переместить несуществующий документ или переместить документ на несуществующую полку*;  
- a – add – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться. *Корректно обработайте ситуацию, когда пользователь будет пытаться добавить документ на несуществующую полку*.

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}

def main ():
    
    print("Перечень команд:",
    "p – команда, которая спросит номер документа и выведет имя человека, которому он принадлежит;",
    "s – команда, которая спросит номер документа и выведет номер полки, на которой он находится;",
    "l – команда, которая выведет список всех документов в формате: passport \"2207 876234\" \"Василий Гупкин\";",
    "as – команда, которая спросит номер новой полки и добавит ее в перечень",
    "d – команда, которая спросит номер документа и удалит его из каталога и из перечня полок;",
    "m – команда, которая спросит номер документа и целевую полку и переместит его с текущей полки на целевую;",
    "a – команда, которая добавит новый документ в каталог и в перечень полок, спросив его номер, тип, имя владельца и номер полки, на котором он будет храниться.",
    "", sep="\n")
    
    while True:
        command = input("Введите команду: ")
        if command.lower() == "p": showname()
        elif command.lower() == "s": showshelf()
        elif command.lower() == "l": showlist()
        elif command.lower() == "as": addshelf()
        elif command.lower() == "d": deldoc()
        elif command.lower() == "m": movedoc()
        elif command.lower() == "a": adddoc()
        elif command.lower() == "q":
            return print("Завершение работы программы.")
        else:
            print("Вы ввели несуществующую команду, повторите ввод команды.")

def showname ():
    numberdoc = input("Введите номер документа: ")
    for document in documents:
        if document["number"] == numberdoc:
            return print("Пользователь документа: ", document["name"])
    return print("Такого номера документа нет в базе.")

def showshelf ():
    numberdoc = input("Введите номер документа: ")
    for document in documents:
        if document["number"] == numberdoc:
            for key, value in directories.items():
                if value.count(numberdoc) > 0:
                    return print("Документ хранится на полке: ",key)
    return print("Такого номера документа нет в базе.")

def showlist ():
    print("Список документов:")
    for document in documents:
        print("passport \"", document["number"], "\" \"", document["name"], "\"", sep="")
    return

def addshelf ():
    numbernewshelf = str(input("Введите номер полки: "))
    for key, value in directories.items():
        if numbernewshelf == key:
            return print("Такой номер полки уже существует.")
        else:
            directories[numbernewshelf] = []
            return #Не понял почему, если писать в одну строку return directories[numbernewshelf] = [] , то ошибка.

def deldoc ():
    numberdoc = input("Введите номер документа: ")
    check = False
    count = 0
    for document in documents:
        if document["number"] == numberdoc:
            del(documents[count])
        count += 1        
    for key, value in directories.items():
        if value.count(numberdoc) > 0:
            del_key = key
            check = True    
    if check:
        del(directories[del_key])
        return
    else:
        return print("Такого номера документа нет в базе.")    

def movedoc ():
    numberdoc = input("Введите номер документа: ")
    check = False
    for key, value in directories.items():
        if numberdoc in value:
            erase_shelf = key
            check = True
    if not check:
        return print("Такого номера документа нет в базе.")    
    numbershelf = str(input("Введите номер полки: "))
    if directories.get(numbershelf) == None:
        return print("Такой полки не существует.")
    tmp_value = directories[erase_shelf]
    tmp_value.remove(numberdoc)
    directories[erase_shelf] = tmp_value    
    tmp_value = directories[numbershelf]
    tmp_value.append(numberdoc)
    directories[numbershelf] = tmp_value    
    return

def adddoc ():
    numberdoc = str(input("Введите номер документа: "))
    typedoc = str(input("Введите тип документа: "))
    owndoc = str(input("Введите имя владельца документа: "))
    numbershelf = str(input("Введите номер полки: "))
    if directories.get(numbershelf) == None:
        return print("Такой полки не существует.")
    newdoc = {}
    newdoc['type'] = typedoc #Ковычки тут и далее в двух строках ставил специально одинарные (это не ошибка стилистики), как в исходных словарях
    newdoc['number'] = numberdoc
    newdoc['name'] = owndoc
    documents.append(newdoc)
    tmp_value = directories[numbershelf]
    tmp_value.append(numberdoc)
    directories[numbershelf] = tmp_value
    return

main()
print("Актуальный перечень документов:", documents, "Актуальный перечень документов, хранимых на полках", directories, sep="\n")
```
