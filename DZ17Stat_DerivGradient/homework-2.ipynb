### Задание 1

Придумайте функции со следующими свойствами (в a) b) c) область определения X любая, какая вам удобна, главное - функция с нужными свойствами):

a) бесконечное количество локальных экстремумов

b) 2 локальных экстремума

с) 3 локальных экстремума

d*) область определения функции - 1 точка

```python
# Функция с бесконечным количеством локальных экстремумов

x_values = []
y_values = []

for x in range(-300, 300):
    x = x / 25
    x_values.append(x)

for x in x_values:
    y_values.append(math.sin(x))

plt.plot(x_values, y_values)
plt.show()
```

```python
# Функция с 2-мя локальными экстремумами

x_values = []
y_values = []

for x in range(-300, 600):
    x = x / 10
    x_values.append(x)

for x in x_values:
    y_values.append(math.pow(x,3)/50 - math.pow(x,2))

plt.plot(x_values, y_values)
plt.show()
```

```python
# Функция с 3-мя локальными экстремумами

x_values = []
y_values = []

for x in range(-300, 300):
    x = x / 10
    x_values.append(x)

for x in x_values:
    y_values.append(50/(math.pow(x,2)+0.5) + math.pow(x,2))

plt.plot(x_values, y_values)
plt.show()
```

```python
# Функция с областью определения функции - 1 точка

x_values = []
y_values = []

for _ in range(-300, 300):
    x = 5
    x_values.append(x)

z = 0
for x in x_values:
    y_values.append(x+z)
    z += 1

plt.plot(x_values, y_values)
plt.show()
```
